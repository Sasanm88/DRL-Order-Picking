# Observe the performance of trained network

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

%matplotlib notebook

def draw_grid(n_row, n_aisle, depot_col):
    n_col = n_aisle * 3 
    fig, ax = plt.subplots(figsize=(10, 10))

    for i in range(n_row):
        for j in range(n_col):
            blue_condition = i == 0 and j == depot_col
            black_condition = j%3!=1 and 0<i and i<n_row-1
            color = 'gray' if black_condition else 'white'
            if blue_condition:
                color = 'blue'
            square = plt.Rectangle((j,i), 1, 1, fc=color, ec='black')
            ax.add_patch(square)

    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim(0, n_col)
    ax.set_ylim(0, n_row)

    return fig, ax

# Function to create a circle

def create_circle(stateM, stateP, circles, texts):
    row, col, room = stateM
    row_ = n_row - 1 - row
    color = 'red'
    circle1 = plt.Circle((col + 0.5, row_ + 0.5), 0.3, fc=color, ec='black')
    ax.add_patch(circle1)
    text_object = ax.text(col + 0.5, row_ + 0.5, str(room.item()), color='black', ha='center', va='center', fontsize=12)
    circles.append(circle1)
    texts.append(text_object)

    for i in range(stateP.shape[0]):
        for j in range(stateP.shape[1]):
            if stateP[i,j] > 0:
                row_ = n_row - 1 - i
                color = 'green'
                circle2 = plt.Circle((j + 0.5, row_ + 0.5), 0.2, fc=color, ec='black')
                ax.add_patch(circle2)
                circles.append(circle2)
                text_object = ax.text(j + 0.5, row_ + 0.5, str(int(stateP[i,j].item())), color='black', ha='center', va='center', fontsize=12)
                texts.append(text_object)


saved_trajectory = []
actions = []
rewards = []
history = []
    
env = Env(n_row, n_aisle, arrival_rate, capacity, depot, inter_aisle_distance, pick_up_time, drop_off_time, alpha)
stateM, stateP, allowed_actions, _, loc, prize = env.reset_show(history, 1)
statesM = [loc]
prizes = [prize]

with torch.no_grad():
    for i in range(n_aisle*500):
        old_stateM = stateM.clone().detach()
        old_stateP = stateP.clone().detach()
        old_allowed_actions = allowed_actions.copy()
        x = agent.target_net.forward(stateM.unsqueeze(0), stateP.unsqueeze(0))[0]
        action = allowed_actions[torch.argmax(x[allowed_actions])]
#         action = agent.target_net.select_optimal_action(stateM, stateP, allowed_actions)
        stateM, stateP, reward, allowed_actions, _, loc, prize, _ = env.step_show(action)
        saved_trajectory.append((old_stateM, old_stateP, old_allowed_actions, x, action, stateM, stateP, reward))
        for i in range(len(loc)):
            statesM.append(loc[i])
            prizes.append(prize[i])
        
fig, ax = draw_grid(n_row, n_aisle, env.depot_col)
circles = []  # List to store the circle objects
texts = []

def update(frame):
    plt.title(f'{frame}/{len(statesM)}')
    for circle in circles:
        circle.remove()  # Remove the previous circles
    for text_object in texts:
        text_object.remove()
    circles.clear()
    texts.clear()
    stateM = statesM[frame]
    stateP = prizes[frame]
    create_circle(stateM, stateP, circles, texts)
    
#     ax.add_patch(circle2)
    
#     circles.append(circle2)
    
anim = FuncAnimation(fig, update, frames=len(statesM), repeat=False, interval=50)

# Display the animation
plt.show()
# anim.save("10_row_5_shelves_arrival_rate=0.1_capacity=10.gif", writer= "pillow")
```
